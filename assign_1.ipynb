```python
import pandas as pd

df = pd.read_csv("Groceries_dataset.csv")  # If it's a .csv file
df.head()

```

```python
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_items = df['itemDescription'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_items.values, y=top_items.index, palette="viridis")
plt.title('Top 10 Most Purchased Grocery Items')
plt.xlabel('Purchase Count')
plt.ylabel('Item')
plt.tight_layout()
plt.show()
```

```python
top_customers = df['Member_number'].value_counts().head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_customers.values, y=top_customers.index, palette="magma")
plt.title('Top 10 Most Active Customers')
plt.xlabel('Number of Purchases')
plt.ylabel('Customer ID')
plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')
monthly_purchases = df.groupby('Month').size()

plt.figure(figsize=(12, 6))
monthly_purchases.plot(kind='line', marker='o')
plt.title('Monthly Purchase Trend')
plt.xlabel('Month')
plt.ylabel('Total Purchases')
plt.grid(True)
plt.tight_layout()
plt.show()
```
